# [좌표 압축](https://www.acmicpc.net/problem/18870)




<!-- ![](image/18870.png) -->



각 숫자보다 작은 숫자들의 개수를 찾는 것이므로  
현재 내가 압축을 해야 하는 number보다 작은 (바로 앞의) 숫자 A의 'A보다 작은 숫자들의 개수'를 알고 있다면 더 쉬워진다. → 기록해두자.

```python
import sys

input = sys.stdin.readline

n = int(input())
ls = list(map(int, input().split()))

# 좌표 압축을 담을 개수
dp = [0]*n

# 정렬시킬 리스트 s
s = []
for i in range(n):
    s.append((i, ls[i])) # 리스트 s에는 인덱스와 좌표가 같이 들어간다.

s.sort(key=lambda x: (x[1], x[0])) # 좌표와 인덱스 순으로 오름차순 정렬

# 가장 작은 숫자가 있을 0번째는 굳이 비교할 이유없이 항상 0이다.
for i in range(1, n): 
    if s[i][1] > s[i-1][1]: # 현재 좌표보다 작은 좌표들의 개수니까 
        dp[s[i][0]] = dp[s[i-1][0]]+1 # 이전 좌표의 개수보다 하나 늘어난 값이 들어간다. 
    else:
        dp[s[i][0]] = dp[s[i-1][0]] # 정렬된 이후라서 else문에는 무조건 앞선 좌표와 같은 경우만 남는다.

print(*dp)
```
