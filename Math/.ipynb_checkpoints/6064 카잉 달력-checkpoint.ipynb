```python
import math
import sys

input = sys.stdin.readline
# year = Ma+x = Nb+y  year 

#       .
T = int(input())
for _ in range(T):
    m, n, x, y = map(int, input().split())
    last = (m*n)//math.gcd(m, n)  # LCM   
    if m == x and n == y:
        print(last)
        continue
    i = 0
    num = 0
    ans = -1
    while num < last:
        num = m*i + x
        if num % n == y % n:
            ans = num
            break
        i += 1
    print(ans)
```
