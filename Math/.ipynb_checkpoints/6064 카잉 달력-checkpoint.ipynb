## [날짜 계산](https://www.acmicpc.net/problem/1476) 문제 푸는 [방식](https://github.com/0general/baekjoon_explanation/blob/main/Brute%20force-%EB%B8%8C%EB%A3%A8%ED%8A%B8%ED%8F%AC%EC%8A%A4/1476.ipynb)과 유사하다.

## 단순하므로 따로 그림은 없다.


## 그러나 연립합동방정식은 이용한다면 더 효율적으로 풀 수 있는 문제이므로
## 공부 후 다시 풀어보도록 하자.

```python
import math
import sys

input = sys.stdin.readline
# year = Ma+x = Nb+y 인 year 찾기

# 연립 일차 합동식 공부해서 다시 해볼 것.
T = int(input())
for _ in range(T):
    m, n, x, y = map(int, input().split())
    last = (m*n)//math.gcd(m, n)  # LCM 최소공배수이자 마지막 해
    if m == x and n == y:
        print(last)
        continue
    i = 0
    num = 0
    ans = -1
    while num < last:
        num = m*i + x
        if num % n == y % n:
            ans = num
            break
        i += 1
    print(ans)
```
