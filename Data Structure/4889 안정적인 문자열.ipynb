# 문자열의 끝까지 다 확인했을 때 남은 여는 괄호의 절반만 닫는 괄호로 바뀌면 됨

```python
import sys

input = sys.stdin.readline
num = 1
while True:
    open = 0  # 남은 여는 괄호의 개수
    temp = 0
    arr = list(input().rstrip())
    if arr[0] == '-':
        break
    for i in range(len(arr)):
        if arr[i] == '{':
            open += 1
        else:
            if open == 0:
                temp += 1  # 변환 횟수
                open += 1  # 여는 괄호로 변환해서 셈
            else:
                open -= 1
    print(f"{num}. {open//2 + temp}")  # 여는 괄호의 절반 변환 + 변환 횟수
    num += 1
```
