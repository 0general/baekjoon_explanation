```python
import sys

input = sys.stdin.readline
num = 1
while True:
    open = 0  #    
    temp = 0
    arr = list(input().rstrip())
    if arr[0] == '-':
        break
    for i in range(len(arr)):
        if arr[i] == '{':
            open += 1
        else:
            if open == 0:
                temp += 1  #  
                open += 1  #    
            else:
                open -= 1
    print(f"{num}. {open//2 + temp}")  #     +  
    num += 1
```
