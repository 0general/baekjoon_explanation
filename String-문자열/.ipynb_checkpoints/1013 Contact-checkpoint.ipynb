# 정규 표현식 

## fullmatch 메소드 : 패턴과 문자열이 남는 부분 없이 완벽하게 일치하는지를 검사

```python
import sys
import re

input = sys.stdin.readline

p = re.compile('(100+1+|01)+')
for _ in range(int(input())):
    s = input().strip()
    if p.fullmatch(s):
        print("YES")
    else:
        print("NO")
```
