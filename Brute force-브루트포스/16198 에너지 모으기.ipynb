```python
import sys
```

```python
def BT(hap, cnt):
    global mx
    if cnt == n-2:
        mx = max(mx, hap)
    for i in range(1, n-1):
        if not visited[i]:
            visited[i] = True # 삭제 처리 대신 visited를 true
            a = 0
            for d in range(i-1, -1, -1): # 삭제된 수들을 지나쳐서 i-1 위치의 숫자 찾기
                if not visited[d]:
                    a = w[d]
                    break
            for u in range(i+1, n): # 삭제된 수들을 지나쳐서 i+1 위치의 숫자 찾기
                if not visited[u]:
                    a *= w[u]
                    break
            BT(hap + a, cnt + 1)
            visited[i] = False
```

## 사실 이 문제는 visited처리를 하는 것보다 실제로 문제처럼 원소를 삭제하는 것이 가장 좋다. 

## visited로 처리했을 때 시간도 메모리도 훨씬 많이 사용하게 된다.

## temp = w[i]로 저장해두고

## visited[i] = True 부분에는 del w[i]를 쓰고,

## visited[i] = False 부분의 코드에는 w.insert(i,temp)를 사용하면 된다.

```python
input = sys.stdin.readline

n = int(input())
w = list(map(int, input().split()))
visited = [False]*(n)
mx = 0
BT(0, 0)
print(mx)
```
