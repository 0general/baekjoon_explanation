```python
def team_make_BT(idx, team):
    if idx >= n or len(team) > n//2:
        return
    team_make_BT(idx+1, team)
    if len(team) < n//2:  #             
        t = team + [idx]
        all_team.append(t)
        team_make_BT(idx+1, t)

n = int(input())
arr = [list(map(int, input().rstrip().split())) for _ in range(n)]

mini = 40000
all_team = []
team_make_BT(0, [])
for team in all_team:
    abl = 0
    link = [i for i in range(n) if i not in team]  #       
    for i in range(len(team)-1):  #     
        for j in range(i+1, len(team)):
            abl += arr[team[i]][team[j]] + arr[team[j]][team[i]]
    for x in range(len(link)-1):  #     
        for y in range(x+1, len(link)):
            abl -= arr[link[x]][link[y]] + arr[link[y]][link[x]]
    abl = abs(abl)
    if abl < mini:
        mini = abl
    if mini == 0:  #  0  
        break

print(mini)
```
