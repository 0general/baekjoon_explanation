```python
# import time
#  : https://zidarn87.tistory.com/339 ->  
#answer1 = [0, 1, 0, 0, 2, 10, 4, 40, 92, 352, 724, 2680, 14200, 73712, 365596]


def check(x, y, num):
    for i in range(1, n-x):
        if 0 <= y-i:
            col[x+i][y-i] += num
        if y+i < n:
            col[x+i][y+i] += num
        col[x+i][y] += num


def BT(x):  #   ,    
    global answer
    if x == n:
        answer += 1
        return
    for i in range(n):
        if col[x][i] == 0:
            check(x, i, 1)
            BT(x+1)  #   
            check(x, i, -1)


n = int(input())
answer = 0

#        n   .
col = [[0]*n for _ in range(n)]
# s = time.time()
BT(0)
# print(time.time()-s)
# input 14  150.74167490005493  .

print(answer)
# pypy3  .
```
