# [N과 M (7)](https://www.acmicpc.net/problem/15656)



<!-- ![](image/15656.png)-->



```python
def BT(cnt, string):
    if cnt == m:
        print(string)
        return
    else:
        for i in range(n):
            BT(cnt+1, string+str(nums[i])+" ")
```

```python
# 자연수 개수, 수열의 길이
n, m = map(int, input().split())
nums = list(map(int, input().split()))
nums.sort() #사전 순으로 증가하는 순서랬으므로 미리 정렬해놓고 시작

BT(0, "")
```
