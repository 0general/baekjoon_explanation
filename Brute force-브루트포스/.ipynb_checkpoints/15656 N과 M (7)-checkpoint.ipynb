# [N M (7)](https://www.acmicpc.net/problem/15656)



<!-- ![](image/15656.png)-->



```python
def BT(cnt, string):
    if cnt == m:
        print(string)
        return
    else:
        for i in range(n):
            BT(cnt+1, string+str(nums[i])+" ")
```

```python
#  ,  
n, m = map(int, input().split())
nums = list(map(int, input().split()))
nums.sort() #      

BT(0, "")
```
