```python
import sys
```

```python
# idx  , cnt  string   ,      .
def BT(idx, cnt, string):
    if len(answer) >= 1:  #              
        return
    if cnt == k:
        answer.append(string)
        return
    if sign[cnt] == '<':
        for j in range(idx+1, 10):  #      
            if visit[j] == 0:
                visit[j] = 1
                BT(j, cnt+1, string+str(j))
                visit[j] = 0
            if len(answer) >= 1:
                break
    else:
        for j in range(idx-1, -1, -1):  #      
            if visit[j] == 0:
                visit[j] = 1
                BT(j, cnt+1, string+str(j))
                visit[j] = 0
            if len(answer) >= 1:
                break
```

```python
input = sys.stdin.readline
k = int(input())
visit = [0]*(10)
sign = input().rstrip().split()

answer = []
for i in range(9, -1, -1): ########   -   .
    if len(answer) == 1:
        break
    visit[i] = 1
    BT(i, 0, str(i))  # number, cnt, string
    visit[i] = 0

print(answer.pop())

for i in range(10):  ########   -   .
    if len(answer) == 1:
        break
    visit[i] = 1
    BT(i, 0, str(i))
    visit[i] = 0

print(answer.pop())
```
