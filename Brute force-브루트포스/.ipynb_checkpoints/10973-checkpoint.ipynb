# 이전 순열

## 규칙
### 1. 숫자 카드의 배열은 작은 수에서 큰 수로 커진다.

### 2. 수가 더 작아지려면 자릿수에 해당하는 숫자의 위치에 더 작은 숫자를 넣어야 한다.

### 3. 사용할 수 있는 숫자와 숫자의 개수가 하나로 정해져있을 때 수가 커지려면 
### &nbsp;&nbsp; 동일 자릿수에 현재 사용 가능한 숫자 중 더 큰 숫자로의 교체만이 가능하다.

![](image/10973.png)

```python
n = int(input())
f = list(map(int, input().split()))  # 숫자 비교 조심

idx = -1
while idx > -n and f[idx] >= f[idx-1]:
    idx -= 1
if idx == -n:
    print(-1)
else:  # 1 6 4 2 3 5
    j = -1
    while f[idx-1] <= f[j]:  # f[idx-1] 와 바꿀 숫자는 f[idx-1]보다 작지만 뒤에서 가장 큰 수
        j -= 1
    f[idx-1], f[j] = f[j], f[idx-1]
    ans = f[:idx] + sorted(f[idx:], reverse=True)
    for i in ans:
        print(i, end=' ')
```
