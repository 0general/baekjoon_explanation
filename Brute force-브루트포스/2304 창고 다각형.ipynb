```python
import sys

input = sys.stdin.readline
```

```python
#   n
n = int(input())
arr = []
for _ in range(n):
    arr.append(tuple(map(int, input().split())))

arr.sort(key=lambda x: x[0])  # x   
mx = 0
id = 0
```

```python
for i in range(n):
    if arr[i][1] > mx: #        
        mx = arr[i][1]
        id = i

a = arr[0]
area = arr[id][1]  #       

#       
for i in range(id):
    if arr[i+1][1] > a[1]:  #   
        area += (arr[i+1][0]-a[0])*a[1]  #  *     
        a = arr[i+1]  #     

#       
a = arr[n-1]
for i in range(n-1, id, -1):
    if arr[i-1][1] >= a[1]:  #     
        #          
        #            
        #         
        #          
        area += (a[0]-arr[i-1][0])*a[1]
        a = arr[i-1]

print(area)
```
