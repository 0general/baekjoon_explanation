## N×N board 위에 N개의 퀸을 놓으려면 모든 행(or 열)에 하나씩 퀸을 선택한다.

## 2022 기준으로 이전의 python 풀이 코드 중에는 재채점 시 시간 초과되는 경우가 있을 것이다.  
정확한 이유는 아직 모르지만, 다른 분들 말씀에 따르면 파이썬 버전 업(백준 서버 파이썬 버전 : 3.9.7 → 3.10.1) 문제일 것이라 추측된다. 


## 각 행(혹은 열)마다 퀸을 선택할 때, 이전 모든 퀸과 비교하는 경우가 위에 해당된다.


## 따라서 방식을 바꾸어 각 행에서 valid check를 할 수 있도록 기록하도록 코드를 작성했다. 

```python
# import time
# 참고 : https://zidarn87.tistory.com/339 -> 시간 초과
#answer1 = [0, 1, 0, 0, 2, 10, 4, 40, 92, 352, 724, 2680, 14200, 73712, 365596]


def check(x, y, num):
    for i in range(1, n-x):
        if 0 <= y-i:
            col[x+i][y-i] += num
        if y+i < n:
            col[x+i][y+i] += num
        col[x+i][y] += num


def BT(x):  # 선택할 행 인덱스이자, 지금까지 선택된 퀸의 개수
    global answer
    if x == n:
        answer += 1
        return
    for i in range(n):
        if col[x][i] == 0:
            check(x, i, 1)
            BT(x+1)  # 다음 행 체크
            check(x, i, -1)


n = int(input())
answer = 0

# 한 행에 반드시 하나를 선택할 수 있어야 n개를 선택할 수 있다.
col = [[0]*n for _ in range(n)]
# s = time.time()
BT(0)
# print(time.time()-s)
# input이 14일 때 150.74167490005493 초가 걸림.

print(answer)
# pypy3로 채점하면 통과된다.
```
