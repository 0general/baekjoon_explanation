# [특정 거리의 도시 찾기](https://www.acmicpc.net/problem/18352)



# 다익스트라 이용!



한 지점에서 다른 곳으로 거리 찾기

<!-- ![](image/18352.png) -->



```python
import heapq
import sys

input = sys.stdin.readline
INF = 300001
N, M, K, X = map(int, input().split())
dist = [INF]*(N+1)
arr = [set() for _ in range(N+1)]
dist[X] = 0
h = []
ans = set()
for _ in range(M):
    a, b = map(int, input().split())
    arr[a].add(b)

heapq.heappush(h, (0, X))
while h:
    c, n = heapq.heappop(h)
    if dist[n] < c: # 이미 더 적은 거리로 갱신된 도시는 굳이 볼 필요없다.
        continue
    if c >= K: # K 거리를 넘으면 더 계산할 필요가 없다.
        continue
    for i in arr[n]:
        t = c+1 # 인접한 도시 간 거리는 1로 모두 동일
        if t < dist[i]:
            dist[i] = t
            if t == K:
                ans.add(i)
            else:
                heapq.heappush(h, (t, i))


if len(ans) == 0:
    print(-1)
else:
    for i in sorted(list(ans)):
        print(i)
```
