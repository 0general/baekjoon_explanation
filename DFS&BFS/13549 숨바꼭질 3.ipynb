# [숨바꼭질](https://www.acmicpc.net/problem/13549)

```python
from collections import deque
```

```python
# 수빈이 위치 n, 동생 k
MAX = 200001
n, k = map(int, input().split())
time = 0
visited = [False]*MAX
q = deque()
ans = 0
```

```python
visited[n] = True # 수빈이 시작 위치
q.append((n, time))
```

```python
while q:
    now, t = q.popleft()
    if now == k:
        print(t)
        break
    x = 2*now
    for x in (now*2, now-1, now+1): # 순간이동은 시간이 걸리지 않으므로 먼저 q에 삽입하도록 for문의 첫 번째로 배치한다.
        if 0 <= x < MAX and not visited[x]:
            visited[x] = True
            if x == now*2:
                q.append((x, t))
            else:
                q.append((x, t+1))
```
