```python
import sys



def dfs():
    global count
    if len(stack) == 0: #   
        return
    now = stack.pop()
    if len(graph[now]) == 0: #   == 
        count += 1
        return
    for i in graph[now]:
        if not visited[i]:
            visited[i] = True
            stack.append(i)
            dfs()


input = sys.stdin.readline

n = int(input())

#  
tree = list(map(int, input().split()))

#   
delete = int(input())

#    (=   )  graph
graph = [[] for _ in range(n)]
visited = [False]*n

stack = []
for i in range(n):
    if i == delete: #   
        continue
    if tree[i] == -1: # i  
        stack.append(i) #  
        visited[i] = True
    else: 
        graph[tree[i]].append(i) # i  graph  i 


#    count 
count = 0
dfs()
print(count)
```
