### 트리의 정의 : 사이클이 없는 그래프
> 즉, 방향성이 있는 그래프처럼 생각하면 된다.   
> 부모 노드는 다음 경로로 자식 노드를 기억해둔다고 생각하자.  
> 리프 노드는 다음 경로가 없는 노드가 된다. 즉 graph[node] = []인 node가 리프노드이다.  


```python
import sys



def dfs():
    global count
    if len(stack) == 0: # 루트 노드가 없으면
        return
    now = stack.pop()
    if len(graph[now]) == 0: # 자식노드가 없으면 == 리프노드
        count += 1
        return
    for i in graph[now]:
        if not visited[i]:
            visited[i] = True
            stack.append(i)
            dfs()


input = sys.stdin.readline

n = int(input())

#부모 노드 받기
tree = list(map(int, input().split()))

# 삭제할 노드 받기
delete = int(input())

# 각 노드의 자식 노드(= 그래프 연결 관계)를 담을 graph
graph = [[] for _ in range(n)]
visited = [False]*n

stack = []
for i in range(n):
    if i == delete: # 삭제할 노드이면 무시
        continue
    if tree[i] == -1: # i가 루트 노드이면
        stack.append(i) # 스택에 담아둠
        visited[i] = True
    else: 
        graph[tree[i]].append(i) # i노드의 부모의 graph에 자식 i를 추가


# 리프노드의 수를 셋 count 변수
count = 0
dfs()
print(count)
```
