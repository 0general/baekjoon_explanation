```python
import sys
from collections import deque
```

```python
input = sys.stdin.readline
n = int(input())  #  
me, find = map(int, input().split())

p = [[-1, []] for _ in range(n+1)]
visited = [False]*(n+1)
for _ in range(int(input())):
    i, j = map(int, input().split())
    p[j][0] = i  # j  i
    p[i][1].append(j)  # i  j

    '''
          x,y   .     x    y   .
                 .
    '''
```

```python
ans = -1
q = deque()
visited[me] = True
q.append((me, 0))
while q:
    now, line = q.popleft()
    if now == find:
        ans = line
        break
    for c in p[now][1]:  #  
        if not visited[c]:
            visited[c] = True
            q.append((c, line+1))
    if p[now][0] != -1 and not visited[p[now][0]]: #     
        visited[p[now][0]] = True
        q.append((p[now][0], line+1))

print(ans)
```
