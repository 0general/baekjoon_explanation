```python
import sys
from collections import deque
```

```python
def bfs():  # h   
    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]
    while q:
        i, j = q.popleft()
        for k in range(4):
            x, y = i+dx[k], j+dy[k]
            if 0 <= x < n and 0 <= y < n and graph[x][y] > h and not visit[x][y]: #        
                visit[x][y] = True
                q.append((x, y))
```

```python
input = sys.stdin.readline

n = int(input())
graph = []

for _ in range(n):
    graph.append(list(map(int, input().split())))
```

```python
q = deque()
ans = 1  #  0     (  1)

for h in range(1, 101):  # 
    visit = [[False]*n for _ in range(n)]
    temp = 0
    for i in range(n):
        for j in range(n):
            if h < graph[i][j] and not visit[i][j]:  #     .
                temp += 1 #     
                visit[i][j] = True
                q.append((i, j))
                bfs() # bfs     visited .
    if temp == 0:  #      0 for 
        break
    ans = max(ans, temp)

print(ans)
```
