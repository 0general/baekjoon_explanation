```python
import sys
from collections import deque
```

```python
input = sys.stdin.readline
n = int(input())  # n 
```

```python
r1, c1, r2, c2 = map(int, input().split())

dx = [-2, -2, 0, 0, 2, 2] #   
dy = [-1, 1, -2, 2, -1, 1]

visited = [[False]*n for _ in range(n)]
q = deque()

visited[r1][c1] = True
```

```python
if r1 == r2 and c1 == c2:
    print(0)
else:
    flag = -1
    q.append((r1, c1, 0))
    while q:
        x, y, cnt = q.popleft()
        for i in range(len(dx)):
            nx, ny = x+dx[i], y+dy[i]
            if 0 <= nx < n and 0 <= ny < n and not visited[nx][ny]:
                visited[nx][ny] = True
                if nx == r2 and ny == c2:  #   ==  
                    flag = cnt + 1
                    break
                q.append((nx, ny, cnt+1))
        if flag > 0:  #   0 
            break
    print(flag)
```
