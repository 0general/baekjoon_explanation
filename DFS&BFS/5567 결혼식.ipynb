# 상근이의 친구의 수와 그 친구들의 친구만 세면 된다.
# 기본 BFS이므로 코드와 주석만으로 설명 가능 (그림 없음)

```python
import sys
from collections import deque
```

```python
input = sys.stdin.readline
```

```python
N = int(input())
graph = [[] for _ in range(N+1)]
visited = [False]*(N+1)
q = deque()
```

```python
answer = 0 # 하객 수

for _ in range(int(input())):
    a, b = map(int, input().split())
    if a == 1:
        visited[b] = True
        q.append(b)  # 친구들이 큐에 모두 들어감 (이 친구들의 친구 수 검사해야하므로)
        answer += 1 # 상근이의 친구 수 하객 처리
    else:
        graph[a].append(b)
        graph[b].append(a)

while q:  # 친구의 친구 하객 처리
    f = q.popleft()
    for i in graph[f]:
        if not visited[i]:
            visited[i] = True
            answer += 1

print(answer)
```
