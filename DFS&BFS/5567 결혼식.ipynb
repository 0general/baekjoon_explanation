#  BFS     ( )

```python
import sys
from collections import deque
```

```python
input = sys.stdin.readline
```

```python
N = int(input())
graph = [[] for _ in range(N+1)]
visited = [False]*(N+1)
q = deque()
```

```python
answer = 0 #  

for _ in range(int(input())):
    a, b = map(int, input().split())
    if a == 1:
        visited[b] = True
        q.append(b)  #     (    )
        answer += 1 #     
    else:
        graph[a].append(b)
        graph[b].append(a)

while q:  #    
    f = q.popleft()
    for i in graph[f]:
        if not visited[i]:
            visited[i] = True
            answer += 1

print(answer)
```
