# [숫자판 점프](https://www.acmicpc.net/problem/2210)



# 브루트 포스, DFS 이용 (기본 내용이므로 그림 없음)

```python
import sys
```

```python
def dfs(i, j, string):
    global answer
    if len(string) == 6:
        if not visited[int(string)]:
            visited[int(string)] = True
            answer += 1
        return
    else:
        for k in [(-1, 0), (1, 0), (0, 1), (0, -1)]:
            x, y = i+k[0], j + k[1]
            if 0 <= x < 5 and 0 <= y < 5: # 한번 지나갔던 곳을 다시 지나가도 되므로 숫자판 내에 있기만 하면 된다.
                dfs(x, y, string+graph[x][y])
```

```python
input = sys.stdin.readline
```

```python
graph = [list(input().rstrip().split()) for _ in range(5)]

visited = [False for i in range(1000000)] # 000000 부터 999999까지
answer = 0 # 숫자의 개수
```

```python
for i in range(5):
    for j in range(5): # 숫자판 위의 모든 곳을 시작 위치로
        dfs(i, j, graph[i][j])

print(answer)
```
