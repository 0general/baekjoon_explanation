```python
#   Ctrl + Z

import sys


def find_sub(start, end):
    if start > end:
        return
    if start == end:
        print(tree[start])
        return
    s, e = start+1, end
    ans = start  #        
    while s <= e:
        mid = (s+e)//2
        if tree[mid] <= tree[start]:
            ans = mid
            s = mid + 1
        else:
            e = mid - 1
    find_sub(start+1, ans)
    find_sub(ans+1, end)
    print(tree[start])


input = sys.stdin.readline
sys.setrecursionlimit(int(1e5))  #     
tree = []

while True:
    try:
        i = int(input())
    except:
        break
    tree.append(i)

find_sub(0, len(tree)-1)
```
