```python
import sys


def check(time, m):
    cnt = m
    for k in arr:
        cnt -= time//k #     k    time        
        if cnt <= 0: #   m   
            return True
    return False


input = sys.stdin.readline
n, m = map(int, input().split())
arr = sorted([int(input()) for _ in range(n)])


#     t
l, r = arr[0], arr[-1]*m
t = r
while l <= r:
    mid = (l+r)//2
    if check(mid, m):
        t = mid
        r = mid - 1
    else:
        l = mid + 1
print(t)
```
