# 이분탐색 

t 시간 내에 모든 심사가 가능한 지 확인하고,  
가능한 t 중에 가장 적은 시간을 이분탐색으로 찾는다.

<!-- ![](image/3079.png) -->



```python
import sys


def check(time, m):
    cnt = m
    for k in arr:
        cnt -= time//k # 한 명을 처리하는 데 k 시간이 걸리는 심사대에서 time 시간 내에 처리할 수 있는 사람 수 빼기
        if cnt <= 0: # 이미 목표치 m명을 다 처리했으므로 넘어감
            return True
    return False


input = sys.stdin.readline
n, m = map(int, input().split())
arr = sorted([int(input()) for _ in range(n)])


# 찾아야 하는 시간의 최솟값 t
l, r = arr[0], arr[-1]*m
t = r
while l <= r:
    mid = (l+r)//2
    if check(mid, m):
        t = mid
        r = mid - 1
    else:
        l = mid + 1
print(t)
```
