### 옮길 필요가 없는 아이들 : 다른 아이들이 자리를 옮김으로 인해 자기 자신의 자리로 갈 수 있는 아이들


즉 옮겨지는 아이들의 수를 최소화하기 위해서는 자리순서(오름차순)을 해치는 아이들 중 가장 적은 수만 찾으면 된다.


결국 순서대로 줄을 세우는 것은 단조 증가 수열을 만드는 것과 다를 바 없다.  
가장 긴 증가하는 부분 수열을 찾고 그 외의 요소들을 재배치해주면 된다.  


### LIS(가장 긴 증가하는 부분 수열)의 길이를 구해서 전체 길이에서 빼준다.

```python
import sys

input = sys.stdin.readline

n = int(input())
ls = []
for _ in range(n):
    num = int(input())
    if len(ls) == 0:
        ls.append(num)
    else:
        if ls[-1] < num:
            ls.append(num)
        else:
            s, e = 0, len(ls)-1
            ans = 0
            while s <= e:
                mid = (s+e)//2
                if ls[mid] >= num:
                    ans = mid
                    e = mid - 1
                else:
                    s = mid + 1
            ls[ans] = num

print(n-len(ls))
```
