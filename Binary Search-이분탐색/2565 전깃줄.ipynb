```python
import sys
```

```python
input = sys.stdin.readline
```

```python
#  
n = int(input())
arr = []
for _ in range(n):
    arr.append(tuple(map(int, input().split())))

arr.sort(key=lambda x: x[0])  #     
```

###       (     )   .

```python
lines = [] 
for i in range(n):
    b = arr[i][1]
    if i == 0:  #       .
        lines.append(b)
    else:
        if b > lines[-1]:  #     .
            lines.append(b)
        else:
            s, e = 0, len(lines)-1
            ans = len(lines)-1
            while s <= e:
                mid = (s+e)//2
                if lines[mid] > b:  #    b        
                    ans = mid
                    e = mid-1
                else:
                    s = mid + 1
            lines[ans] = b  #    .
            #          ,      .

print(n-len(lines))  #         .
```
