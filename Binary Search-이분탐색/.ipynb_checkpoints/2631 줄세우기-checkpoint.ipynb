### LIS(    )     .

```python
import sys

input = sys.stdin.readline

n = int(input())
ls = []
for _ in range(n):
    num = int(input())
    if len(ls) == 0:
        ls.append(num)
    else:
        if ls[-1] < num:
            ls.append(num)
        else:
            s, e = 0, len(ls)-1
            ans = 0
            while s <= e:
                mid = (s+e)//2
                if ls[mid] >= num:
                    ans = mid
                    e = mid - 1
                else:
                    s = mid + 1
            ls[ans] = num

print(n-len(ls))
```
