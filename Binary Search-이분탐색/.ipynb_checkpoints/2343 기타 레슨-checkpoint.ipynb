# [기타 레슨](https://www.acmicpc.net/problem/2343)



<!-- ![](image/2343.png) -->


```python
n, m = map(int, input().split())
nums = list(map(int, input().split()))

l, r = max(nums), 1000000000  # 하나의 동영상은 하나의 블루레이 안에 들어갈 수 있어야 한다. 10000 * 100000
ans = 0 # 구하려는 건 블루레이의 크기
while l <= r:
    mid = (l+r)//2 # 크기 찾기 시작
    temp = 0 # mid 크기를 가진 블루레이 안에 들어갈 수 있는 영상들의 크기 합
    group = 0 # 블루레이 개수
    for i in range(n):
        if (temp + nums[i]) > mid:  # 블루레이 안에 들어갈 수 없으면
            temp = 0 # temp reset
            group += 1 # 블루레이 개수 추가
        temp += nums[i]
    if group >= m:  # 필요한 블루레이 개수가 많다면
        l = mid + 1 # 블루레이 크기를 키워야 함
    else:
        ans = mid # 일단 가능함
        r = mid-1 # 더 줄이기 가능한지 확인
print(ans)
```
