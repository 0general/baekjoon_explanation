```python
n, m = map(int, input().split())
nums = list(map(int, input().split()))

l, r = max(nums), 1000000000  #         . 10000 * 100000
ans = 0 #    
while l <= r:
    mid = (l+r)//2 #   
    temp = 0 # mid          
    group = 0 #  
    for i in range(n):
        if (temp + nums[i]) > mid:  #     
            temp = 0 # temp reset
            group += 1 #   
        temp += nums[i]
    if group >= m:  #    
        l = mid + 1 #    
    else:
        ans = mid #  
        r = mid-1 #    
print(ans)
```
