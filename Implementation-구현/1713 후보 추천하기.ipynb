```python
import sys


#     
def check(x, i):
    if x in ans:
        increase_num(x)
    else:
        post_photo(x, i)


#    
def post_photo(x, i):
    if len(ans) == n:
        d = find_minimum()
        del ans[d]
    ans[x] = [1, i]  #   1, i 


#     
def increase_num(x):
    ans[x][0] += 1


#     
def find_minimum():
    return min(ans.keys(), key=(lambda x: ans[x]))


input = sys.stdin.readline


n = int(input())
num = int(input())

rec = list(map(int, input().split()))
ans = {}

for i, p in enumerate(rec):
    check(p, i)

print(*sorted(ans.keys()))
```
