```python
from collections import deque
import sys
```

```python
input = sys.stdin.readline
```

```python
n = int(input())
board = [[0]*n for _ in range(n)]
visited = [[False]*n for _ in range(n)]
```

```python
for _ in range(int(input())): #   
    i, j = map(int, input().split())
    i -= 1
    j -= 1
    board[i][j] = 1
```

```python
dir = [[0, 1], [-1, 0], [0, -1], [1, 0]]  # 

dt = [0]*(10000+1)  # idx 
```

####     

```python
for _ in range(int(input())):
    x, c = input().rstrip().split()  #  x 3  c 4   .
    if c == 'L':  #  
        dt[int(x)] = 1
    else:
        dt[int(x)] = -1

for i in range(1, len(dt)):  #   
    dt[i] = (dt[i-1]+dt[i]) % 4
```

```python
i = 0
q = deque()
q.append((0, 0))
visited[0][0] = True
while True:
    x, y = q[-1]
    k = i if i < len(dt) else len(dt)-1
    nx, ny = x+dir[dt[k]][0], y + dir[dt[k]][1]
    if 0 <= nx < n and 0 <= ny < n and not visited[nx][ny]:
        visited[nx][ny] = True
        q.append((nx, ny))
        if board[nx][ny] == 0:  #  
            visited[q[0][0]][q[0][1]] = False
            q.popleft()
        else:
            board[nx][ny] = 0
    else:
        break
    i += 1

print(i+1)
```
