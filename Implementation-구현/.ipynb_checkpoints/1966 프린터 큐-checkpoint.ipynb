```python
from collections import deque
import heapq
import sys
```

##    

```python
input = sys.stdin.readline
t = int(input())
```

```python
for _ in range(t):
    #   ,   
    n, m = map(int, input().split())
    # n 
    docs = list(map(int, input().rstrip().split()))
    q = deque([i for i in range(n)])
    h = []
    cnt = 0
    for d in docs:
        heapq.heappush(h, -d)
    while q:
        now = q.popleft()  # index
        if docs[now] == -h[0]:  #    
            cnt += 1
            heapq.heappop(h)
            if now == m:
                break
        else: #      
            q.append(now)
    print(cnt)
```
