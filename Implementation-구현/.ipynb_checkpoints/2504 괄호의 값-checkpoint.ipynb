# [괄호의 값](https://www.acmicpc.net/problem/2504)



<!-- ![](image/2504.png) -->



```python
def solution(string):
    stack = []
    pair = {')': '(', ']': '['} # 짝 지정
    num = {')': 2, ']': 3} # 숫자
    l = 0

    if string[-1] in {'(', '['}:  # 여는 괄호로 끝나는 경우는 처음부터 배제
        return 0
    if len(string) % 2 == 1:
        return 0
    for s in string:
        if s == '(' or s == '[':  # 여는 괄호
            stack.append(s)
            l += 1  # 여는 괄호의 숫자
        else:  # 닫는 괄호가 나올 때마다 처리해 줄 것이기에 stack에는 늘 숫자나 여는 괄호만 남아있다.
            c = pair[s]  # ) 이면 (, ] 이면 [
            if len(stack) == 0:
                return 0
            if stack[-1] == c:  # 닫는 괄호의 짝
                stack.pop()
                stack.append(num[s])
                l -= 1
            elif str(stack[-1]).isdigit(): #닫는 괄호의 바로 앞이 숫자일때는
                a = 0
                while str(stack[-1]).isdigit(): # 여는 괄호를 만날때까지 그 안의 숫자들 합산 {type() == int 을 쓰는게 낫다}
                    a += stack.pop()
                    if len(stack) == 0:  # 여는 괄호가 없이 종료 → 닫는 괄호가 짝을 만나지 못함
                        return 0
                if stack[-1] == c:  # 괄호의 올바른 짝을 만나면
                    stack.pop()
                    stack.append(num[s]*a) #계산해서 숫자를 넣어줌
                    l -= 1 # 남아있는 여는 괄호 수
                else:  # 숫자들을 처리하고 처음 만난 괄호가 닫는 괄호와 짝이 맞지 않을 때
                    return 0
            else:  # 닫는 괄호와 짝이 맞지 않을 때
                return 0
    if l != 0: # 아직 처리되지 않은 여는 괄호의 수가 남아있다면
        return 0
    return sum(stack)  # 숫자만 남아있는 상태
```

```python
string = input()
print(solution(string))
```
