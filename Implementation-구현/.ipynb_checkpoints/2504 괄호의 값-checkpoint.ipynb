```python
def solution(string):
    stack = []
    pair = {')': '(', ']': '['} #  
    num = {')': 2, ']': 3} # 
    l = 0

    if string[-1] in {'(', '['}:  #      
        return 0
    if len(string) % 2 == 1:
        return 0
    for s in string:
        if s == '(' or s == '[':  #  
            stack.append(s)
            l += 1  #   
        else:  #        stack     .
            c = pair[s]  # )  (, ]  [
            if len(stack) == 0:
                return 0
            if stack[-1] == c:  #   
                stack.pop()
                stack.append(num[s])
                l -= 1
            elif str(stack[-1]).isdigit(): #    
                a = 0
                while str(stack[-1]).isdigit(): #        {type() == int   }
                    a += stack.pop()
                    if len(stack) == 0:  #          
                        return 0
                if stack[-1] == c:  #    
                    stack.pop()
                    stack.append(num[s]*a) #  
                    l -= 1 #    
                else:  #           
                    return 0
            else:  #      
                return 0
    if l != 0: #       
        return 0
    return sum(stack)  #   
```

```python
string = input()
print(solution(string))
```
