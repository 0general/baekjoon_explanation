# [A  B](https://www.acmicpc.net/problem/16953)



 BFS       **** 

# A B    B A  

#    !

```python


A, B = map(int, input().split())


cnt = 1
while A <= B:
    if A == B:
        print(cnt)
        break
    if B % 2 == 0:
        B //= 2
    elif B % 10 == 1:  #     1  
        B //= 10
    else:
        print(-1)
        break
    if A > B:
        print(-1)
        break
    cnt += 1
```
