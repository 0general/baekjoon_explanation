```python
import sys
```

```python
input = sys.stdin.readline
n = int(input())
nums = list(map(int, input().split()))
```

```python
dp = [num for num in nums] # i         
mx = dp[0]
```

```python
for i in range(1, n):
    for j in range(i-1, -1, -1): # i index   
        if nums[j] < nums[i]:
            dp[i] = max(dp[i], dp[j] + nums[i])
            if dp[j] == mx:
                mx = dp[i]
                break
    if dp[i] > mx:
        mx = dp[i]

print(mx)
```
