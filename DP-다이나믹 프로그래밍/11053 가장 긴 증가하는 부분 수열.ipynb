# [가장 긴 증가하는 부분 수열](https://www.acmicpc.net/problem/11053)

```python
import sys

input = sys.stdin.readline
n = int(input())  # 수열의 크기
nums = list(map(int, input().split()))
```

```python
# i번째 수가 선택되었을 때  가장 긴 증가하는 부분 수열
dp = [1]*n 
```

```python
# 수열의 어떤 원소도 반드시 선택된다는 가정이 없고 수열이 크기 순으로 정렬되어 있지 않으므로 
# 수열 길이의 최대값을 저장할 mx를 따로 지정한다.
mx = 1  
```

```python
for i in range(1, n):
    for j in range(i):
        if nums[i] > nums[j]:  # 무조건 자기자신을 마지막으로 두는 경우
            dp[i] = max(dp[i], dp[j]+1) 
        if dp[i] > mx: # 수열에서 고려되는 숫자가 i번째 수로 단 하나 추가되었을 뿐이므로, mx+1이 되면 남은 것들은 고려할 필요 없다.
            mx = dp[i]
            break

print(mx)
```
