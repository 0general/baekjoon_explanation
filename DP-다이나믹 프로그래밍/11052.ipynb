[카드 구매하기-최대값 구하기](https://www.acmicpc.net/problem/11052)

<!-- ![카드 구매하기](image/11052.png) -->


<!-- ![예시](image/11052-ex.png) -->


```python
n = int(input())
dp = [0] + list(map(int, input().split()))

for now in range(2, n+1):
    for i in range(1, now//2 + 1):
        dp[now] = max(dp[now], dp[i] + dp[now-i])

print(dp[n])
```
