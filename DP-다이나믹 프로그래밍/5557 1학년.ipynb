# [1](https://www.acmicpc.net/problem/5557)



# dp[idx][k] : idx    k(0~20)    

<!-- ![](image/5557.png) -->


```python
import sys

input = sys.stdin.readline

n = int(input())  # 3 <= n <= 100       -> dp

#        dp count  
dp = [[0]*21 for _ in range(n-1)]

ls = list(map(int, input().split()))

last = ls[-1]
dp[0][ls[0]] = 1

for i in range(1, n-1):  #       
    num2 = ls[i]
    for num1, x in enumerate(dp[i-1]):  #         
        if x != 0:  #  num1      x
            temp = num1+num2  # i     
            if 0 <= temp <= 20:
                dp[i][temp] += x  # i-1    num1 num2    
            temp = num1-num2  # i     
            if 0 <= temp <= 20:
                dp[i][temp] += x  # i-1    num1 num2    

print(dp[-1][last])
```
