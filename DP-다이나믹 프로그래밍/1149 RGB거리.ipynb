```python
import sys
input = sys.stdin.readline
```

```python
ans = [0, 0, 0]  #   , ,     
```

##            .

```python
for i in range(int(input())):
    temp = list(map(int, input().rstrip().split()))
    temp[0] += min(ans[1], ans[2])
    temp[1] += min(ans[0], ans[2])
    temp[2] += min(ans[0], ans[1])
    ans = temp

print(min(ans))
```
