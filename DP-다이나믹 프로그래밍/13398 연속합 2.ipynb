```python
import sys
```

```python
input = sys.stdin.readline

n = int(input())
nums = list(map(int, input().split()))
dp = [[num, 0] for num in nums]
mx = dp[0][0]
```

```python
for i in range(1, n):
    dp[i][0] = max(dp[i-1][0]+nums[i], nums[i])  #        
    #         (    ,   )
    dp[i][1] = max(dp[i-1][1]+nums[i], dp[i-1][0])
    mx = max(mx, dp[i][0], dp[i][1])

print(mx)
```
