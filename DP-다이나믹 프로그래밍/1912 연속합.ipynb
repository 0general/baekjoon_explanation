# [연속합](https://www.acmicpc.net/problem/1912)

  

<!-- ![](image/1912.png) -->


```python
import sys
```

```python
input = sys.stdin.readline

n = int(input())
dp = list(map(int, input().split()))
```

```python
# 해당 숫자를 선택했을 때 최대값
answer = dp[0]
for i in range(1, n):
    dp[i] = max(dp[i-1]+dp[i], dp[i]) # 누적합보다 자기 자신이 더 크다면 자기 자신 선택
    if dp[i] > answer:
        answer = dp[i]

print(answer)
```
