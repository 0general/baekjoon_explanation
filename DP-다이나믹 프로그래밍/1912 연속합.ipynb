```python
import sys
```

```python
input = sys.stdin.readline

n = int(input())
dp = list(map(int, input().split()))
```

```python
#     
answer = dp[0]
for i in range(1, n):
    dp[i] = max(dp[i-1]+dp[i], dp[i]) #        
    if dp[i] > answer:
        answer = dp[i]

print(answer)
```
