```python
import sys
```

```python
def BT(num):  #  Binary Search
    if len(stack) == 0:
        stack.append(num)
        return
    start, end = 0, len(stack)-1
    ans = len(stack)
    while start <= end:  # num  
        mid = (start+end)//2
        if stack[mid] <= num:  # num      
            ans = mid
            end = mid-1
        else:  # num     
            start = mid + 1
    if ans == len(stack):
        stack.append(num)
    else:
        stack[ans] = num
    return
```

```python
input = sys.stdin.readline

n = int(input())
seq = list(map(int, input().split()))
stack = []
for num in seq:
    BT(num)

print(len(stack))
```
