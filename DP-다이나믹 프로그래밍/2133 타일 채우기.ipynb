```python
n = int(input())
```

# dp[i] 
## i  [    ,     ,  2  ]

---

#   
### -   i-2   ( 3)   
### -   i-1  1   ( 1,  1)   

---

# i        
### -  i-1      ,     2 
### -  i-1 2    2   

---

#     
### -   i-1         

```python
dp = [[0, 0, 0] for i in range(n+1)]
dp[0] = [1, 0, 0]
dp[1] = [0, 2, 0]
for i in range(2, n+1):
    dp[i][0] = dp[i-2][0] + dp[i-1][1]
    dp[i][1] = dp[i-1][0]*2+dp[i-1][2]
    dp[i][2] = dp[i-1][1]

print(dp[n][0])
```
