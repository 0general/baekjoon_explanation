```python
import sys
```

```python
sys.setrecursionlimit(25000)  # 250000     
```

#      DFS 

```python
def dfs(i, j):
    if i == m-1 and j == n-1: 
        return 1
    else:
        if dp[i][j] == -1:
            dp[i][j] = 0
            for k in range(4):
                x, y = i+dx[k], j+dy[k]
                if 0 <= x < m and 0 <= y < n and arr[x][y] < arr[i][j]:
                    dp[i][j] += dfs(x, y) # i,j        
    return dp[i][j]
```

```python
input = sys.stdin.readline
dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]
m, n = map(int, input().split())
arr = [list(map(int, input().split())) for _ in range(m)]
```

```python
dp = [[-1]*n for _ in range(m)]
dp[m-1][n-1] = 1 #   . dfs if return 1  
```

```python
print(dfs(0, 0))
```
