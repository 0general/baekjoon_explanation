# 경로를 찾는 것이라서 그래프 탐색임은 너무 명확
# 하지만 최소 경로를 구하는 것이 아니며, 
# 조건에 따라 동일한 경로가 중첩되기도 하므로 이 문제는 DP를 사용해야 한다.

# DP의 정의 
   
## dp[i][j] = (i, j)부터 목적지(m-1, n-1)까지의 경로의 수


# dp에서 고려해야 할 상태 
## - 현 위치 i,j에서 인접한 4개의 칸 (북, 서, 남, 동 방향)
## - 각 칸의 값이 현 위치의 값보다 작아야 이동 가능

```python
import sys
```

```python
sys.setrecursionlimit(25000)  # 250000으로 메모리 초과 나서 하나 줄여봄
```

# 하나의 경로를 제대로 찾는게 중요하므로 DFS 이용

```python
def dfs(i, j):
    if i == m-1 and j == n-1: 
        return 1
    else:
        if dp[i][j] == -1:
            dp[i][j] = 0
            for k in range(4):
                x, y = i+dx[k], j+dy[k]
                if 0 <= x < m and 0 <= y < n and arr[x][y] < arr[i][j]:
                    dp[i][j] += dfs(x, y) # i,j를 거쳐서 목적지에 도달한 모든 경우를 합산하여 저장하게 됨
    return dp[i][j]
```

```python
input = sys.stdin.readline
dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]
m, n = map(int, input().split())
arr = [list(map(int, input().split())) for _ in range(m)]
```

```python
dp = [[-1]*n for _ in range(m)]
dp[m-1][n-1] = 1 # 사실 초기화 필요없다. dfs의 if문에서 return 1을 해주기 때문에
```

```python
print(dfs(0, 0))
```
