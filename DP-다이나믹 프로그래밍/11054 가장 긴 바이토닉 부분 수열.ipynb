# [가장 긴 바이토닉 부분 수열](https://www.acmicpc.net/problem/11054)

  

```python
n = int(input())  # 수열의 크기
num = list(map(int, input().split()))

increase = [1 for _ in range(n)] # 증가하는 수열의 최대 길이
decrease = [1 for _ in range(n)] # 감소하는 수열의 최대 길이

for i in range(n): # 증가 수열과 감소 수열을 미리 구해놓는다.
    for j in range(i):
        if num[i] > num[j]:
            increase[i] = max(increase[j]+1, increase[i])
        if num[-i-1] > num[-j-1]:
            decrease[-i-1] = max(decrease[-j-1]+1, decrease[-i-1])

result = [increase[i] + decrease[i]-1 for i in range(n)] # 자기자신이 증가수열과 감소수열에 각각 1번씩 총 2번 세어졌으므로 -1
print(max(result))
```
