```python
n = int(input())
m = int(input())
```

# dp[i] = [a,b,c]  i     
## a =   ,  b=   i  , c =   
#  i       
#       .  
##  
## i     i+1        

```python
dp = [[0, 0, 0] for _ in range(n+1)]
```

```python
for i in range(m):
    num = int(input())
    dp[num][0], dp[num][2] = -1, -1

dp[0] = [0, 1, -1]
for i in range(1, n):
    if dp[i][0] != -1:  #    
        dp[i][0] = dp[i-1][2] if dp[i-1][2] != -1 else 0
        dp[i][2] = dp[i-1][1] if dp[i+1][2] != -1 else 0
        if dp[i-1][2] != -1:  #   i   i-1    .
            dp[i][1] += dp[i-1][2]
    dp[i][1] += dp[i-1][1]

if dp[n-1][2] != -1 and dp[n][0] != -1:
    print(dp[n-1][1] + dp[n-1][2])
else:
    print(dp[n-1][1])
```
