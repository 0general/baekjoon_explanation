## 길이가 아닌 합을 저장하기만 한다는 점만 다르지, 동일한 문제이다.

```python
import sys
```

```python
input = sys.stdin.readline
n = int(input())
nums = list(map(int, input().split()))
```

```python
dp = [num for num in nums] # i번째 숫자를 마지막 수로 하는 증가수열 중 가장 큰 합
mx = dp[0]
```

```python
for i in range(1, n):
    for j in range(i-1, -1, -1): # i index 이전까지 모두 검사
        if nums[j] < nums[i]:
            dp[i] = max(dp[i], dp[j] + nums[i])
            if dp[j] == mx:
                mx = dp[i]
                break
    if dp[i] > mx:
        mx = dp[i]

print(mx)
```
