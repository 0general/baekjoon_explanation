```python
from math import sqrt
```

```python
from collections import deque
```

```python
def solution(n):
    x = int(sqrt(n))

    #    : n    
    dp = [0]*(n+1)
    for i in range(1, x+1):
        dp[i**2] = 1  # .    1

    for i in range(2, n+1):
        if dp[i] == 0:  #    
            a = int(sqrt(i))
            #    12 = 3^2 + (1^2)*3  4 .   (4^2)*3 3
            dp[i] = dp[i-a**2] + 1
            for k in range(1, a):
                if dp[i] == 1 or dp[i] == 2:
                    break
                dp[i] = min(dp[i], dp[i-k**2]+1)
    print(dp[n])


solution(int(input()))
```

---

```python
def solution2(n):  # using queue -    
    ls = []
    q = deque()
    dp = [0]*(n+1)
    for i in range(1, int(sqrt(n))+1):
        a = i**2
        ls.append(a) #  
        q.append(a) #    
        dp[a] = 1

    while q:
        now = q.popleft()
        for i in ls:
            next = now + i #  
            if next > n:
                break
            if not dp[next]:
                dp[next] = dp[now]+1
                q.append(next) #   
    print(dp[n])


solution2(int(input()))
```
