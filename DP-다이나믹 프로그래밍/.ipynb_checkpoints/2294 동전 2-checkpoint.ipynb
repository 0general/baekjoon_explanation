```python
import sys
from collections import deque
```

```python
input = sys.stdin.readline
N, K = map(int, input().split())
dp = [0]*(100000+1)  # index     
#     dp K+1    .   
coin = []
q = deque()
```

```python
for i in range(N):
    number = int(input())
    #                .
    if dp[number] == 0:
        coin.append(number)
        dp[number] = 1
        q.append(number)

coin.sort()  #          

while q:
    now = q.popleft()
    for c in coin:
        temp = now + c
        if temp <= K and dp[temp] == 0:  #           .
                                          #          dp 0    X
            dp[temp] = dp[now]+1
            q.append(temp)
if dp[K] == 0:
    print(-1)  #   .      
else:
    print(dp[K])
```
