# [가장 긴 감소하는 부분 수열](https://www.acmicpc.net/problem/11722)

  

## DP가 아닌 이분 탐색으로 풀어보자

## 이분 탐색을 사용하는 방식

## - 숫자가 들어갈 위치를 찾고 그 위치의 숫자를 대체한다.

## - 내림차순 수열의 모양은 계속 달라지지만, 중요한 것은 숫자 자체가 아닌 숫자의 대소비교다.

## - 따라서 숫자의 크고 작음의 비교만 가능해진다면 길이를 구하는 방식엔 달라지는 것이 없다.

```python
import sys
```

```python
def BT(num):  # 내림차순 Binary Search
    if len(stack) == 0:
        stack.append(num)
        return
    start, end = 0, len(stack)-1
    ans = len(stack)
    while start <= end:  # num의 위치 찾기
        mid = (start+end)//2
        if stack[mid] <= num:  # num이 크거나 같으면 더 앞쪽으로 가야 함
            ans = mid
            end = mid-1
        else:  # num이 더 작으면 뒤쪽 그룹에서 찾기
            start = mid + 1
    if ans == len(stack):
        stack.append(num)
    else:
        stack[ans] = num
    return
```

```python
input = sys.stdin.readline

n = int(input())
seq = list(map(int, input().split()))
stack = []
for num in seq:
    BT(num)

print(len(stack))
```
